```python
saved = open("pickle/game_reviews.pickle","rb")
game_reviews = pickle.load(saved)
game_reviews.head()
```

```python
user_game = open("pickle/user_games.pickle","rb")
user_games = pickle.load(user_game)
user_games.shape
```

```python
user_games = user_games.rename(columns={'user_id': 'user', 'item_id': 'item'})
user_games_list = user_games[['item', 'user']]
user_games_list.head()
```

```python
reviews = game_reviews[['item_id', 'user_id','review']]
reviews = reviews.rename(columns={'user_id': 'user', 'item_id': 'item'})
reviews['item'].nunique()
```

```python
rev_item = set(reviews['item'])   ## reviewed item  3682
user_item = set(user_games['item'])  ##items in user item data  10978
item_with_rev = user_item.intersection(rev_item) ## items that have reviews
user_item_rev = user_games_list[user_games_list['item'].isin(list(item_with_rev))] 
result = pd.merge(user_item_rev, reviews, how = 'outer', on=['item', 'user'])
```

```python
def groupby_count(df, group, count):
    game_count = pd.DataFrame()
    game_count['count'] = df.groupby(group)[count].nunique()
    return game_count
```

```python
def prune(df, condition):     ## returns a dataframe that meet the given condition
    user_n = df.loc[df['count'] < condition ]
    return user_n
```

```python
game_count = groupby_count(result, 'user', 'item')
game_count.sort_values(by = 'count', ascending = False).head()
```

```python
user_5 = prune(game_count, 5)
user_less_5 = user_5.index
pruned_data_5 = result.set_index('user').drop(user_less_5)
pruned_data_5.reset_index(inplace = True)
pruned_data_5.head()
```

```python
pruned_data_5.isna().sum()
```

```python
pairs_user = list(sample_users(pruned_data_5, 5, 10000, xf.SampleN(1)))
```

```python
import pandas as pd
import numpy as np
import pickle
from lenskit import batch, topn, util

file = open("comp_user.pickle","rb")
pairs_user = pickle.load(file)

truth = pd.concat((p.test for p in pairs_user))
result = pd.read_parquet('results/steam/comp/recommendations.parquet')
def RR(rec, truth):
    #recs = pd.read_parquet(file_name)
    rla = topn.RecListAnalysis()
    rla.add_metric(topn.recall)
    RR_result = rla.compute(rec, truth)
    return RR_result

RR_algo_comp = RR(result, truth)
print("done with recip_rank")

##RR_algo_comp.to_parquet('results/steam/comp/lt_recall.parquet')
#RR_algo_comp.to_csv("results/steam/pruned_5_new/RR_algo1000.csv", ignore_index=True)

pickle_out = open("results/steam/comp/lt_recall.pickle","wb")
pickle.dump(RR_algo_comp, pickle_out, protocol = 4)
pickle_out.close()


### join with algo name
legend = pd.read_csv("results/steam/comp/runs.csv")
legend = legend.set_index('RunId').loc[:,'AlgoStr']

RR_algo = RR_algo_comp.join(legend, on='RunId')

#RR_algo.to_parquet('results/steam/comp/lt_recal100.parquet')
#RR_algo.to_csv("results/steam/pruned_5_new/RR_algoname1000.csv", ignore_index=True)

pickle_out = open("results/steam/comp/lt_RR_algo.pickle","wb")
pickle.dump(RR_algo, pickle_out, protocol = 4)
pickle_out.close()

print("done with merging")
```
